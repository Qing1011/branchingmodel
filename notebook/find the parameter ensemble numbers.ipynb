```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat, savemat
import pandas as pd
import scipy.special as SS
import scipy.stats as SSA
import copy
import random
import math
from sklearn.model_selection import ParameterGrid
import os
import numpy.linalg as LA
import gzip

# load pickle module
import pickle
```

```python
rs = np.arange(0.01, 1, 0.02)

R0s = np.arange(1.5, 6.5, 0.1)

param_grid = {'R0': R0s, 'r' : rs}

grid = ParameterGrid(param_grid)

para_dict = list(grid)
```

```python
def check_file_exists(filename):
    """Check if the file exists.
    if exist return True, else return False
    """
    try:
        f = open(filename, 'r')
        f.close()
        return True
    except FileNotFoundError:
        return False
```

```python
file_dir = '../../../../../../Volumes/My Passport/branching/results_ic_300/'
```

```python
file_dir1 = '../../../../../../Volumes/My Passport/branching/results_ic_200/'
```

```python
para_times_left = []
for i in para_dict[300:500]:
    R0 = i['R0']
    r =  i['r']
    file_name = file_dir+"NewInf_R0-{}_r-{}.npy.gz" .format(np.round(R0,2),np.round(r,2))
    if check_file_exists(file_name):
        para_times_left.append((R0,r,200))
    else:
        file_name1 = file_dir1+"NewInf_R0-{}_r-{}.npy.gz" .format(np.round(R0,2),np.round(r,2))
        a = check_file_exists(file_name1)
        if a:
            para_times_left.append((R0,r,300))
        else:
            para_times_left.append((R0,r,500)) 
```
