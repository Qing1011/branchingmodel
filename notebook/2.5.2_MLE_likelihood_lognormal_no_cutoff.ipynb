```python
def load_gzipped_numpy(filename):
    try:
        with gzip.open(filename, 'rb') as f:
            return np.load(f, allow_pickle=True)
    except FileNotFoundError:
        return [0]
```

```python
rs = np.array([20, 10, 2.0, 1.0, 0.5, 0.2, 0.1, 0.05, 0.025])
```

```python
len_r = len(rs)
```

```python
Ensembles = np.zeros((len(rs),300,3142,60))
```

```python
R0 = 2.5
export_dir = '/Users/qingyao/Documents/branching_data/simulation/python_cutoff_addno/'
for r_idx in range(len_r):
#     print(r_idx)
    r = rs[r_idx]
    sub_export_dir = export_dir+'branching_R0-{}_r-{}/'.format(np.round(R0,2),np.round(r,3))
    for i in range(300):
        export_names = sub_export_dir+'NewInf_R0-{}_r-{}_{}.npy.gz'.format(np.round(R0,2),np.round(r,3),(i+1))
        temp = load_gzipped_numpy(export_names)
        if len(temp) == 1:
            print(r, i)
        else:
            Ensembles[r_idx,i,:,:] = temp
```

```python
Ensembles_0 = copy.deepcopy(Ensembles)
Ensembles_0[Ensembles_0 == 0] = 0.000000000000001
Ensembles_log = np.log(Ensembles_0)
```

```python
from matplotlib.ticker import ScalarFormatter
formatter = ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-5,5))
plt.figure(figsize=(4.5, 5),dpi=450)
for l_idx in range(16):
    l = locations[l_idx]
    print(l)
    log_data = Ensembles_log[5,:,l,59]
    mu = np.mean(log_data)
    sigma = np.std(log_data)
    # The shape parameter for the lognormal distribution is the standard deviation of the log-transformed data
    shape = sigma
    # The scale parameter for the lognormal distribution is e^mu
    scale = np.exp(mu)
    
    temp2 = Ensembles[5,:,l,59]
    
    axs_i = plt.subplot(4,4,l_idx+1)
    
    plt.hist(temp2[temp2>1],density=True,color='grey',bins=30)
    
    d_statistic_norm, p_value_norm = kstest(temp2, 'norm',args=(np.mean(temp2), np.std(temp2)))
    print('normal:', d_statistic_norm, p_value_norm )
    d_statistic_log, p_value_log = kstest(temp2, 'lognorm', args=(mu,sigma))
    print('lognormal:', d_statistic_log, p_value_log )
    
    x_left = np.min(temp2[temp2>1])
    x_right = np.max(temp2[temp2>1])
    x = np.arange(x_left,x_right,10)
    
    y = SSA.lognorm.pdf(x, s=sigma, scale=scale,loc=0)
    plt.plot(x,y,label='lognormal',lw=1)

    y2 = SSA.norm.pdf(x, scale=np.std(temp2),loc=np.mean(temp2))
    plt.plot(x,y2,label='normal',lw=1,color='red')
    
    axs_i.set_xticklabels(axs_i.get_xticks(), rotation=45)
    axs_i.xaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    axs_i.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    axs_i.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
    axs_i.tick_params(axis='both', which='major', labelsize=7) 
    axs_i.xaxis.offsetText.set_fontsize(7)
    axs_i.yaxis.offsetText.set_fontsize(7)

plt.tight_layout()

    
#     plt.title('before log, loc = {}'.format(l))
# plt.legend()
plt.savefig('lognormal_R0-2.5_r-0.2.pdf')
```
