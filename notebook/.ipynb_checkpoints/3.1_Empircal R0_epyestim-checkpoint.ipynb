```python
import scipy.special as SS
import scipy.stats as SSA
Z = 3
Zb = 1
D = 5
Db = 1
latency_p = SSA.gamma.pdf(x=range(18), a=Z, scale=Zb)
infectious_p = SSA.gamma.pdf(x=range(18),a=D, scale=Db)
# v = np.random.random_sample(z_num)
delay_days = latency_p+0.5*infectious_p  # 3+5*0.5
delay_days_pdf = delay_days/np.sum(delay_days)
```

```python
# Generate samples
z_num = 10000000
latency_p = gamma.rvs(a=Z, scale=Zb, size=z_num)
infectious_p = gamma.rvs(a=D, scale=Db, size=z_num)
v = np.random.random_sample(z_num)
delay_days_samples = latency_p + v * infectious_p
```

```python
np.max(delay_days_samples)
```

```python
fig = plt.figure( figsize=(6,3))
plt.bar(range(18),si_distrb,width=1,label='paper',alpha=0.7,color='red')
# plt.bar(range(18),delay_days,width=1)
# plt.bar(range(18),delay_days_pdf,width=1,alpha=0.5)
values, bin_edges, _ = plt.hist(delay_days_samples,density=True, bins=28,alpha=0.5,label='mine',color='blue')
```

```python
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
```

```python
my_si_dis = values/np.sum(values)
```

```python
dataframes_to_concat = []
for fip_i in range(len(fips_main)): #len(fips_main)
    cases_i = cases_sorted.iloc[fip_i,11:356].values
    col_name = cases_sorted.iloc[fip_i,0]
    temp = col_name-84000000
    temp2 = df_fips_sorted.iloc[fip_i].values
    if temp != temp2[0]:
        print(fip_i, temp, temp2[0])
    daily_cases = np.diff(cases_i) 
    daily_cases[daily_cases<0] = 0
    df_cases = pd.DataFrame(daily_cases, index=dates, columns=[col_name])
    df_cases.index.name = 'Date'
#     print(fip_i)
    if df_cases[col_name].max()==0:
        df_Rmean.loc[dates.index, column_name] = 0
    else:
        ch_time_varying_r = covid19.r_covid(df_cases[col_name],smoothing_window=14, r_window_size=7,gt_distribution=my_si_dis)
        column_name = str(temp) + '_R_mean'
        df_Rmean_i = pd.DataFrame(index=dates)
        df_Rmean_i.loc[ch_time_varying_r.index, column_name] = ch_time_varying_r['R_mean']
        dataframes_to_concat.append(df_Rmean_i)
```

```python
# df_Rmean = df_Rmean.copy()
```

```python
df_Rmean = pd.concat(dataframes_to_concat, axis=1)
```
