```python
def find_best_learning_rates(file_dir, x_number, num_splits, num_runs, num_learning_rates):
    # Initialize a matrix to store the best learning rates for each split
    best_learning_rates = np.zeros(num_splits)

    for split in range(num_splits):
        # Initialize a matrix to store the validation losses for each run and learning rate
        val_losses = np.zeros((num_runs, num_learning_rates))

        for lr_idx in range(num_runs * num_learning_rates):
            lr_pos = lr_idx % num_learning_rates
            run = lr_idx // num_learning_rates

            with h5py.File(f'{file_dir}regression_{x_number}/res_{split}_{lr_idx}.hdf5', 'r') as f:
                val_loss = f['val_mse'][()]
                val_losses[run, lr_pos] = val_loss[-1]

        # Find the learning rate with the best average performance across runs for this split
        avg_val_loss_per_lr = np.mean(val_losses, axis=0)
        best_lr_idx = np.argmin(avg_val_loss_per_lr)
        best_learning_rates[split] = best_lr_idx

    return best_learning_rates

# Example usage
file_dir = 'path/to/your/files/'
x_number = 'your_x_number'
num_splits = 1  # Number of splits
num_runs = 10  # Number of runs per split
num_learning_rates = 7  # Number of learning rates

best_lrs = find_best_learning_rates(file_dir, x_number, num_splits, num_runs, num_learning_rates)
print("Best learning rates for each split:", best_lrs)
```

```python
import scipy.stats as stats
def bootstrap_median_confidence_interval(data, ci=95, n_bootstraps=10000):
    bootstrapped_medians = []
    n = len(data)

    for _ in range(n_bootstraps):
        sample = np.random.choice(data, size=n, replace=True)
        bootstrapped_medians.append(np.median(sample))

    lower_bound = np.percentile(bootstrapped_medians, (100 - ci) / 2)
    upper_bound = np.percentile(bootstrapped_medians, 100 - (100 - ci) / 2)
    medians = np.percentile(bootstrapped_medians, 50)

    return medians, lower_bound, upper_bound
```

```python
class GCN(torch.nn.Module):  # the simpliest model that GNN and it is classical, used as baseline
    def __init__(self, num_node_features):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(num_node_features, 128)
        self.conv2 = GCNConv(128, 64)
        self.conv3 = GCNConv(64, 16)
        self.conv4 = GCNConv(16, 8)
        self.fc = torch.nn.Linear(8, 1)

    """
        hyperparameters:
        - number of hidden layers
        - number of hidden channels
        - dropout rate (now it's zero)
        - learning rate <- most important to tune
        - weight decay
        - etc etc.
    """

    def forward(self, data):
        x, edge_index, edge_weight, batch = data.x, data.edge_index, data.edge_attr, data.batch

        x = self.conv1(x, edge_index, edge_weight)
        x = F.elu(x)

        x = self.conv2(x, edge_index, edge_weight)
        x = F.elu(x)

        x = self.conv3(x, edge_index, edge_weight)
        x = F.elu(x)

        x = self.conv4(x, edge_index, edge_weight)
        x = F.elu(x)

        x = global_mean_pool(x, batch)
        x = self.fc(x)

        return x
```

```python
def load_gzipped_numpy(filename):
    try:
        with gzip.open(filename, 'rb') as f:
            return np.load(f, allow_pickle=True)
    except FileNotFoundError:
        return [0]
```

```python
def test(model, testloader, device):
    model.eval()
    predictions = []
    total_loss = 0
    true_values = []
    with torch.no_grad():
        for data in testloader:
            data = data.to(device)
            output = model(data)
            loss = F.mse_loss(output, data.y)
            total_loss += loss.item() * data.num_graphs
            predictions.append(output.cpu())
            true_values.append(data.y.cpu())
    return torch.cat(predictions, dim=0), torch.cat(true_values, dim=0), total_loss / len(testloader.dataset)
```

```python
testloader = DataLoader(dataset, batch_size=128, shuffle=True)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
predictions, y_true, test_mse = test(model, testloader, device)
```
